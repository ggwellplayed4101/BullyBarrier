## 1. Setup and Dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Load clean data

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')
df.info()
```

## 3. View label distribution

```python
sns.countplot(x='label', data=df)
plt.title('Label Distribution')
plt.show()

print(df['label'].value_counts())
```

## 4. Check text length distribution

```python
df['text_length'] = df['text'].apply(len)

sns.histplot(df['text_length'], bins=30)
plt.title('Text Length Distribution')
plt.xlabel('Number of characters')
plt.ylabel('Frequency')
plt.show()
```

## 5. View sample texts per label

```python
for label in sorted(df['label'].unique()):
    print(f"\nLabel {label} examples:")
    display(df[df['label'] == label]['text'].sample(3, random_state=42))
```

## 6. Summary and Findings
- Dataset has 674 samples with no missing values.
- Labels are imbalanced: about 550 zeros and 100 ones.
- Text lengths mostly range from 20 to 120 characters.
- Class imbalance should be addressed in modeling.
- Text lengths are generally short, making preprocessing simpler.
- Next step: clean and preprocess text for modeling.
